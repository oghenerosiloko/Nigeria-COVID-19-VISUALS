```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
covid19=pd.read_csv("COVID_AFRICA.csv")
```

```
covid19.head()
```

```
covid19.shape
```

```
covid19.columns
```

```
#Renaming the columns
covid19.columns=["Country", "Total_cases", "Active_cases", "Recovered_cases", "Total_deaths"]
```

```
covid19.head()
```

```
covid19.info()
```

```
#Cleaning of the data
#Remove commas
covid19['Total_cases']= covid19['Total_cases'].str.replace(',','')
covid19['Active_cases']= covid19['Active_cases'].str.replace(',','')
covid19['Recovered_cases']= covid19['Recovered_cases'].str.replace(',','')
covid19['Total_deaths']= covid19['Total_deaths'].str.replace(',','')
```

```
covid19.head()
```

```
covid19.info()
```

```
#Converting the columns from objects to Numeric values
covid19['Total_cases']=pd.to_numeric(covid19['Total_cases'])
covid19['Active_cases']=pd.to_numeric(covid19['Active_cases'])
covid19['Recovered_cases']=pd.to_numeric(covid19['Recovered_cases'])
covid19['Total_deaths']=pd.to_numeric(covid19['Total_deaths'])
```

```
covid19.info()
```

```
#Addition of death rate and recovery rate columns
covid19["Death_rate"]= round(covid19['Total_deaths']/covid19['Total_cases']*100, 2)
covid19["Recovery_rate"]= round(covid19['Recovered_cases']/covid19['Total_cases']*100, 2)
```

```
covid19.head()
```

```
#drop rows with missing data
covid19.dropna( inplace=True)
```

```
covid19.tail(10)
```

```
covid19.head(2)
```

```
import seaborn as sns
```

```
corelation= covid19.corr()
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```

```
sns.pairplot(covid19)
```

```
sns.relplot(x="Death_rate", y="Recovery_rate", data=covid19)
sns.relplot(x='Total_cases', y='Active_cases', data=covid19)
```

```
sns.displot(covid19["Death_rate"], bins=10)
sns.displot(covid19["Recovery_rate"], bins=10)
```

```
fig_dims= (8, 15)
fig, ax=plt.subplots(figsize=fig_dims)
sns.barplot(x="Total_cases", y="Country",ax=ax, data=covid19)
```

```
fig_dims= (8, 15)
fig, ax=plt.subplots(figsize=fig_dims)
sns.barplot(x="Death_rate", y="Country",ax=ax, data=covid19)
```

```
fig_dims= (8, 15)
fig, ax=plt.subplots(figsize=fig_dims)
sns.barplot(x="Recovery_rate", y="Country",ax=ax, data=covid19)
```

```
covid19.head(1)
```

```
top10death= covid19.nlargest(10, ['Total_deaths'])
fig_dims= (8,8)
fig, ax=plt.subplots(figsize=fig_dims)
sns.barplot(x='Total_deaths', y='Country', data=top10death)
plt.title("Top 10 African Countries with the highest number of Covid-19 Deaths", fontsize= 15)
```

```
top10recovery= covid19.nlargest(10, ['Recovery_rate'])
fig_dims= (8,8)
fig, ax=plt.subplots(figsize=fig_dims)
sns.barplot(x='Recovery_rate', y='Country', data=top10recovery)
plt.title("Top 10 African Countries with the highest Recovery rate of Covid-19", fontsize= 15)
```

```
top10deathrate= covid19.nlargest(10, ['Death_rate'])
fig_dims= (8,8)
fig, ax=plt.subplots(figsize=fig_dims)
sns.barplot(x='Death_rate', y='Country', data=top10deathrate)
sns.set_style=('darkgrid')
plt.title("Top 10 African Countries with the highest Death rate of Covid-19", fontsize= 15)
```

```
top10cases= covid19.nlargest(10, ['Total_cases'])
fig_dims= (8,8)
fig, ax=plt.subplots(figsize=fig_dims)
sns.barplot(x='Total_cases', y='Country', data=top10cases)
plt.title("Top 10 African Countries with the highest number of confirmed Covid-19 Cases", fontsize= 15)
```

```
top10active= covid19.nlargest(10, ['Active_cases'])
fig_dims= (8,8)
fig, ax=plt.subplots(figsize=fig_dims)
sns.barplot(x='Active_cases', y='Country', data=top10active)
plt.title("Top 10 African Countries with the highest number of Active Covid-19 Cases", fontsize= 15)
```
